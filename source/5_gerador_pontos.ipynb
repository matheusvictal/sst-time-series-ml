## Checagem
Precisamos chegar se um ponto aleatório gerado é válido

```python
def check_valid(df, lat, lon):
    ts = SSTHelper.get_sst_series(df, lat, lon).sst.to_list()
    if len(ts) == 0 or pd.isna(ts[0]):
        return False
    return True
```

```python
check_valid(df, 0, -72) # o valor retornado deve ser falso
```

```python
check_valid(df, -22, -72) # o valor retornado deve ser verdadeiro
```

```python
def generate_even():
    return randint(-90, 90) * 2
```

## Geraçao de pontos

```python
list_points = set()
list_not_points = set()
n_points = 0
lat = lon = 0

while n_points < 100:
    lat = generate_even() 
    lon = generate_even()
    while (lat, lon) not in list_not_points and not check_valid(df, lat, lon) and (lat, lon) not in list_points:
        list_not_points.add((lat, lon))
        lat = generate_even()
        lon = generate_even()
    list_points.add((lat, lon))
    n_points += 1
    print(list_points)
```

```python
list_points = list(list_points)
check_valid(df, list_points[0][0], list_points[0][1])
```

```python
points_df = pd.DataFrame(list_points)
```

```python
points_df.rename({0: 'lat', 1: 'lon'}, axis=1).to_csv('../dados/pontos.csv', index=False)
```

## Reler os pontos salvos em csv

```python
def read_list_points(filename):
    list_points = []
    with open(filename, 'r') as file:
        for line in file:
            line_split = line.split(',')
            try:
                list_points.append((int(line_split[0]), int(line_split[1])))
            except ValueError:
                pass
    return list_points
```

```python
list_points = read_list_points('../dados/pontos.csv')
```
